```python
import pandas as pd

# Load the dataset
file_path = '/content/soci_econ_country_profiles.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# Box Plot for GDP per capita
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['GDP per capita (current US$)'])
plt.title('Box Plot of GDP per capita')
plt.show()
```

```python
# Word Cloud for country names
wordcloud = WordCloud(width=800, height=400, background_color ='white').generate(' '.join(data['country']))
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Country Names')
plt.show()
```

```python

# Regression Plot between GDP per capita and Population density
plt.figure(figsize=(10, 6))
sns.regplot(x='Population density (per km2, 2017)', y='GDP per capita (current US$)', data=data)
plt.title('Regression Plot between Population Density and GDP per capita')
plt.show()
```

```python
from wordcloud import WordCloud

# Generate a word cloud for the "Region" column
region_text = ' '.join(data['Region'])

# Create the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(region_text)

# Display the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Regions')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the box plot for GDP per capita across different regions
plt.figure(figsize=(14, 8))
sns.boxplot(x='Region', y='GDP per capita (current US$)', data=data)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.title('Box Plot of GDP per Capita by Region')
plt.xlabel('Region')
plt.ylabel('GDP per Capita (current US$)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the box plot for Sex Ratio across different regions
plt.figure(figsize=(14, 8))
sns.boxplot(x='Region', y='Sex ratio (m per 100 f, 2017)', data=data)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.title('Box Plot of Sex Ratio by Region')
plt.xlabel('Region')
plt.ylabel('Sex Ratio (m per 100 f, 2017)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='Life expectancy at birth, total (years)', y='GDP per capita (current US$)', data=data)
plt.title('Regression Plot: Life Expectancy vs. GDP per Capita')
plt.xlabel('Life Expectancy at Birth (years)')
plt.ylabel('GDP per Capita (current US$)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='Surface area (km2)', y='GDP per capita (current US$)', data=data)
plt.title('Regression Plot: Surface Area vs. GDP per Capita')
plt.xlabel('Surface Area (km2)')
plt.ylabel('GDP per Capita (current US$)')
plt.show()
```
